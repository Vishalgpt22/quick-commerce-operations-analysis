```python
# Vartical stack
vstack_array=np.vstack([p,q])
print(vstack_array)
print(vstack_array.ndim)
print(vstack_array.shape)
```

```python
# Horizontal stack
hstack_array=np.hstack([p,q])
print(hstack_array)
print(hstack_array.ndim)
print(hstack_array.shape)
```

```python
# Concatenate it is also same as Horizontal stack
Concatenate_array=np.concatenate([p,q])
print(Concatenate_array)
print(Concatenate_array.ndim)
print(Concatenate_array.shape)
```

## Statistics

```python
import statistics as sts
```

```python
# mean
x=[25,41,12,32]

print(np.mean(x))
print(sts.mean(x))
```

```python
# Median
y=[12,25,32,41]

print(np.median(y))
print(sts.median(y))
```

```python
# Mode

z=[12,15,25,28,25,100,12,52,25]
print(sts.mode(z))
```

```python
range_example=[25,12,20,10,5,100,78,64,93]
print(range_example)
```

```python
highest=np.max(range_example)
lowest=np.min(range_example)

print(highest)
print(lowest)

print("Range:",highest-lowest)
```

```python
# standard devaiation
std_example=[10,20,30,40,50]
print(std_example)
```

```python
print(np.std(std_example))
print(sts.stdev(std_example))
print("-"*25)

# variance:square of std

print(np.var(std_example))
print(sts.variance(std_example))
```

```python
# Quartile 
iqr_example=[12,4,8,16,20,15,7,6,41,100,37]
print(iqr_example)
```

```python
q3=np.percentile(iqr_example,75)
print("q3:",q3)

q1=np.percentile(iqr_example,25)
print("q1:",q1)

print("IQR:",q3-q1)
```
